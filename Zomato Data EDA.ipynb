# Observation : Zomato Transcation is More in india 

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# Observation 
#1. when rating is 4.1 to 4.9 food is-----> Excellent
#2. wnen rating is 4.0 to 4.4 food is-----> very good
#3. when rating is 3.5 to 3.9 food is----->good

```python
ratings.head()
```

```python
import matplotlib
plt.rcParams["figure.figsize"] = (12,6)
sns.barplot(x = "Aggregate rating", y ="Rating Count",data = ratings )
sns.barplot(x = "Aggregate rating", y ="Rating Count",hue='Rating color',data = ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','green','green'])
```
